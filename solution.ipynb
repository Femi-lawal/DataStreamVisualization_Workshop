````python
import pandas as pd
import psycopg2
import matplotlib.pyplot as plt
from IPython.display import clear_output, display
import time
import os

# üîê Neon DB Config
DB_CONFIG = {
    "host": "ep-twilight-poetry-ad4klocb-pooler.c-2.us-east-1.aws.neon.tech",
    "database": "neondb",
    "user": "neondb_owner",
    "password": "npg_w8YtZHyuePS7",
    "port": "5432",
    "sslmode": "require",
}


class StreamingSimulator:
    def __init__(self, csv_path, delay=2):
        if not os.path.exists(csv_path):
            raise FileNotFoundError(f"‚ùå File not found: {csv_path}")
        
        self.data = pd.read_csv(csv_path)
        self.delay = delay
        self.index = 0
        self.columns = [col for col in self.data.columns if "Axis" in col]
        self.history = pd.DataFrame(columns=self.columns)
        print(f"‚úÖ Loaded {len(self.data)} rows from {csv_path}")
        
        self._drop_and_create_table()

    def _connect(self):
        return psycopg2.connect(**DB_CONFIG)

    def _drop_and_create_table(self):
        """Drop and recreate the robot_readings table from scratch."""
        with self._connect() as conn:
            with conn.cursor() as cur:
                # Drop the table if it exists
                cur.execute("DROP TABLE IF EXISTS robot_readings;")
                # Define the column structure
                column_defs = ", ".join([f'"{col}" FLOAT' for col in self.columns])
                # Create the table
                cur.execute(f"""
                    CREATE TABLE robot_readings (
                        id SERIAL PRIMARY KEY,
                        timestamp TIMESTAMPTZ DEFAULT NOW(),
                        {column_defs}
                    );
                """)
                conn.commit()
        print("üîÅ robot_readings table dropped and recreated.")

    def fetch_latest_data(self):
        """Fetch only the most recent data point and add to history."""
        with self._connect() as conn:
            with conn.cursor() as cur:
                cols = ", ".join([f'"{col}"' for col in self.columns])
                cur.execute(f"""
                    SELECT {cols}
                    FROM robot_readings
                    ORDER BY timestamp DESC
                    LIMIT 1;
                """)
                row = cur.fetchone()

        if row:
            new_df = pd.DataFrame([row], columns=self.columns)
            self.history = pd.concat([self.history, new_df], ignore_index=True)

            # Maintain a sliding 3600-second window
            if len(self.history) > 3600:
                self.history = self.history.iloc[-90:]

            return self.history.copy()
        else:
            return pd.DataFrame(columns=self.columns)        

    def nextDataPoint(self):
        """Read next data point, insert into DB, and return DataFrame."""
        if self.index >= len(self.data):
            print("‚úÖ End of stream.")
            return None

        row = self.data.iloc[self.index]

        # üõ† Convert NumPy float types to native Python float for SQL insertion
        values = [float(row[col]) if pd.notnull(row[col]) else None for col in self.columns]

        # Insert into DB
        with self._connect() as conn:
            with conn.cursor() as cur:
                placeholders = ", ".join(["%s"] * len(values))
                cols = ", ".join([f'"{col}"' for col in self.columns])
                cur.execute(
                    f"INSERT INTO robot_readings ({cols}) VALUES ({placeholders});", values
                )
                conn.commit()

        self.history.loc[len(self.history)] = values
        if len(self.history) > 90:
            self.history = self.history.iloc[-90:]  # 90s window

        self.index += 1
        return self.history.copy()

    def stream_and_plot(self, max_points=300):
        for _ in range(max_points):
            df = self.nextDataPoint()
            if df is None:
                break
            df = self.fetch_latest_data()
            if df.empty:
                print("No data fetched from DB.")
                continue
            self.plot(df)
            time.sleep(self.delay)

    def plot(self, df):
        clear_output(wait=True)
        plt.figure(figsize=(14, 6))
        numeric_df = df[self.columns].apply(pd.to_numeric, errors='coerce').fillna(0)
        # Create reversed time labels: [‚àí(n‚àí1), ..., ‚àí2, ‚àí1, 0]
        time_labels = list(range(-(len(numeric_df) - 1), 1))
        plt.stackplot(time_labels, numeric_df.T.values, labels=self.columns, alpha=0.7)
        plt.title("Realtime Current (Amps, stacked)", fontsize=14)
        plt.xlabel("Time (seconds ago)", fontsize=12)
        plt.ylabel("Current (Amps)", fontsize=12)
        plt.ylim(0, 60)
        plt.legend(loc="upper left", fontsize=9, ncol=2)
        plt.grid(True, linestyle="--", alpha=0.6)
        plt.tight_layout()
        display(plt.gcf())
        plt.close()


csv_file_path = "./data/RMBR4-2_export_test.csv"

# üöÄ Instantiate and run
ss = StreamingSimulator(csv_path=csv_file_path, delay=2)
ss.stream_and_plot(max_points=3600)
````
